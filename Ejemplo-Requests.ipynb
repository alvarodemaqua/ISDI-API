# Peticiones GET

Método utilizado para recibir recursos específicos de una API

```python
url = 'https://dattebayo-api.onrender.com'
endpoint = '/characters'
query_params = { 'page' : 1, 'limit': 1}
respuesta = requests.get(url + endpoint, verify=False)
segunda_respuesta = requests.get(url + endpoint, params=query_params, verify=False)
```

```python
respuesta.status_code
```

```python
respuesta.json()
```

```python
segunda_respuesta.status_code
```

```python
segunda_respuesta.json()
```

```python
json_stored = segunda_respuesta.json()
json_stored
```

```python
json_stored[('characters')]
```

```python
for character in json_stored['characters']:
    print(character['name'])
```

```python
try:
    url = 'https://api.jikan.moe/v4'
    endpoint = '/top/anime'
    query_params = { 'type': 'tv', 'filter': 'upcoming' }
    respuesta_animes = requests.get(url + endpoint, params=query_params, verify=False)
    respuesta_animes.raise_for_status()
except requests.exceptions.HTTPError as httpError:
    print ("Http Error:", httpError)
except requests.exceptions.ConnectionError as conError:
    print ("Error Connecting:", conError)
except requests.exceptions.Timeout as timeError:
    print ("Timeout Error:", timeError)
except requests.exceptions.RequestException as err:
    print ("OOps: Something Else", err)
```

```python
respuesta_animes.json()
```

```python
response = respuesta_animes.json()['data']
```

```python
response
```

```python
import pandas as pd
df = pd.json_normalize(response)
```
